функція .describe() повертає наступні характеристики:
- count (загальна кількість)
- mean (середнє значення)
- std (середнє квадратичне відхилення)
- min (мінімальне значення)
- 25% (перший квантиль, значення більше за 25%)
- 50% (другий квантиль, медіана)
- 75% (третій квантиль,значення більше за 75%)
- max (максимальне значення)

###Створення нового масиву даних (лише з 10-ма рядками завантаженого масиву)

```python
cut_brain = brain.iloc[59:69, :]
cut_brain
```

####Додавання нового стовпчику

```python
new_brain = ["gr1","gr1","gr1","gr1","gr1","gr2","gr2","gr2","gr2","gr2"]
cut_brain.insert(2, "Group", new_brain)
cut_brain
```

####Виведення статистики по всім змінним, розділивши масив даних

```python
cut_brain.groupby('Group').describe()
```
