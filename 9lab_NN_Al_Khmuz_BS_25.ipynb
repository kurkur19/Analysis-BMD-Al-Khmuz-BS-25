##*Теоретична частина*

***Transfer Learning (передача навчання)*** – це методика машинного навчання, де
модель, розроблена для одного завдання, використовується як вихідний пункт
для моделі, розробленої для іншого (зазвичай пов'язаного) завдання. Це
відбувається завдяки перевикористанню основної частини моделі та адаптації
окремих частин під конкретне завдання.


Основні принципи передачі навчання:

- *Використання попереднього знання*: Transfer Learning дозволяє моделі
використовувати знання, отримане під час навчання на одному завданні,
для покращення продуктивності на іншому завданні.
- *Економія ресурсів*: Навчання моделі з нуля може бути дорого коштовним
як у часовому, так і у відношенні до обчислювальних ресурсів.
Застосування передачі навчання може скоротити ці витрати.


**Відмінності між Transfer Learning і традиційним навчанням.**


- Вихідний пункт навчання: У традиційному навчанні модель навчається "з
нуля" на конкретному датасеті. У передачі навчання модель використовує
знання, отримане з інших датасетів або завдань, як вихідний пункт.

- Час та обчислювальні ресурси: Transfer Learning може значно зменшити час
та ресурси, необхідні для навчання моделі, оскільки він використовує
попередньо навчені ваги як вихідний пункт.

- Доступність даних: Традиційне навчання може потребувати великої
кількості даних для ефективного навчання, особливо для складних
моделей. У передачі навчання можливо досягти хороших результатів навіть
із меншими датасетами, так як модель вже "знає" загальні особливості з
інших завдань.

- Застосування: Традиційне навчання зазвичай застосовується, коли є
достатньо даних для навчання моделі та немає доступу до попередньонавчених моделей. Transfer Learning використовується, коли є попередньо
навчені моделі для подібних завдань або коли даних не вистачає для
навчання з нуля.

- Універсальність: Хоча Transfer Learning в основному асоціюється з
глибоким навчанням та згортковими мережами, він також може бути
застосований до інших типів моделей та алгоритмів машинного навчання.

**Основні стратегії передачі навчання: Feature extraction та Fine-tuning.**

- Feature Extraction (видобуток ознак)
***Feature Extraction** – це процес використання вже навчених моделей, особливо їх
початкових шарів, для видобутку корисних ознак з нових даних. Потім ці ознаки
можуть бути використані для навчання нового класифікатора.


**Особливості Feature Extraction:**
- Використовується архітектура попередньо навченої моделі, але фінальний
(або фінальні) шари замінюються новими, які ініціалізуються випадковим
чином.

- Тільки нові шари піддаються навчанню, в той час як ваги попередньо
навчених шарів заморожуються.

- Ця стратегія ефективна, коли доступний обмежений набір даних.


***Fine-tuning (точне налаштування)***

**Fine-tuning** включає в себе налаштування ваг попередньо навченої моделі,
навчаючись на новому датасеті. Тобто, замість заморожування ваг попередньо
навченої моделі, ми дозволяємо їм змінюватися під час навчання.

**Особливості Fine-tuning:**

- Може включати в себе налаштування всіх шарів моделі або тільки
декількох останніх.

- Ця стратегія зазвичай вимагає більшого набору даних, ніж Feature
Extraction, щоб уникнути перенавчання.

- Як правило, навчання починається з менших швидкостей навчання,
оскільки великі зміни ваг можуть "розбити" попередньо навчені ознаки.

- Це може бути особливо корисно, коли новий датасет дуже відрізняється від
датасету, на якому модель була навчена спочатку.


Докладніше, стратегія, яку слід обрати (Feature Extraction або Fine-tuning),
залежить від конкретної задачі, доступної кількості даних та ресурсів для
навчання.


**Переваги Transfer Learning:**
- Швидше навчання: Коли використовуються попередньо навчені моделі,
треба менше часу на досягнення прийнятної точності порівняно з
навчанням моделі з нуля.

- Вимагає менше даних: Transfer learning може бути особливо корисним,
коли доступний тільки обмежений набір даних.

- Вища стійкість до перенавчання: Якщо базова модель була навчена на
великому наборі даних, вона може забезпечити загальні ознаки, які
допоможуть уникнути перенавчання на менших наборах даних.

- Доступність: Існує багато попередньо навчених моделей у загальному
доступі, які можуть бути використані для Transfer Learning.


**Обмеження передачі навчання:**
- Не завжди корисний: Якщо джерельний і цільовий датасети занадто
відрізняються, передача навчання може не принести бажаного ефекту.

- Попередній досвід: Інколи попередньо навчена модель може внести певний
"спотворений" вигляд даних, який може бути не доцільним для нової
задачі.

- Обчислювальні витрати: Деякі попередньо навчені моделі можуть бути
великими і вимагати значних обчислювальних ресурсів, навіть якщо вони
не повністю використовуються.

- Гнучкість: Ви обмежені архітектурою попередньо навченої моделі, хоча і
можна модифікувати її кінцеві шари.

###Етап 2: Попередня обробка даних

На цьому етапі ми нормалізуємо значення пікселів до діапазону [0, 1],
використаємо one-hot кодування для міток класів та застосуємо трансформації для збільшення розмаїття тренувального набору.

```python
# Діагностика: друк форми та значень
print(f"Форма augmented_images: {augmented_images.shape}")
print(f"Максимальне значення пікселів: {augmented_images.max()}")
print(f"Мінімальне значення пікселів: {augmented_images.min()}")

# Перевіряємо тип даних
print(f"Тип даних: {augmented_images.dtype}")
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import cifar100
```

```python
# Завантаження CIFAR-100
(x_train, y_train), (x_test, y_test) = cifar100.load_data(label_mode="fine")

# Масштабування пікселів
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# One-hot кодування міток
y_train = to_categorical(y_train, 100)
y_test = to_categorical(y_test, 100)

# Аугментація даних
datagen = ImageDataGenerator(
    rotation_range=15,      # Випадкове обертання
    width_shift_range=0.1,  # Горизонтальне зміщення
    height_shift_range=0.1, # Вертикальне зміщення
    horizontal_flip=True    # Горизонтальне дзеркалювання
)

# Підготовка генератора
datagen.fit(x_train)

# Генерація batch зображень
augmented_images = next(datagen.flow(x_train, batch_size=9))

# Перевірка форми даних
print(f"Форма augmented_images: {augmented_images.shape}")  # Очікується (9, 32, 32, 3)

# Функція для візуалізації
def plot_augmented_images(images):
    plt.figure(figsize=(10, 10))
    for i in range(len(images)):  # Перегляд усіх зображень у batch
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i])  # Зображення у форматі RGB
        plt.axis('off')
    plt.show()

# Візуалізація
plot_augmented_images(augmented_images)
```

###Етап 3: Моделювання

####Крок 1: Завантаження попередньо натренованої моделі

Ми скористаємося ResNet50, яка добре підходить для класифікації зображень.

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
```

```python
# Завантажуємо ResNet50 без верхніх шарів
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Додаємо власні шари для класифікації CIFAR-100
x = base_model.output
x = GlobalAveragePooling2D()(x)  # Усереднюємо просторові ознаки
x = Dense(256, activation='relu')(x)  # Повнозв'язний шар
predictions = Dense(100, activation='softmax')(x)  # Класифікація на 100 класів

# Створюємо фінальну модель
model = Model(inputs=base_model.input, outputs=predictions)

# Заморожуємо базову модель (всі шари ResNet50)
for layer in base_model.layers:
    layer.trainable = False

# Компіляція моделі
model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Перегляд структури моделі
model.summary()
```

У цій роботі ми використовуємо модель ResNet50 як база для витягання ознак. Попередньо натреновані ваги на ImageNet залишаються незмінними, щоб скористатися їхньою здатністю ефективно витягувати ознаки зі зображень. Для адаптації моделі до задачі класифікації CIFAR-100 були додані:
1. Шар GlobalAveragePooling2D для зменшення розмірності.
2. Повнозв'язний шар із 256 нейронами для додаткової обробки ознак.
3. Кінцевий шар із 100 нейронами, що відповідає кількості класів у датасеті CIFAR-100.

Модель містить 24,137,956 параметрів, із яких 550,244 — тренувальні (нові шари), а решта залишаються замороженими.


####Крок 2: Навчання моделі

Навчимо модель із застосуванням аугментованих даних:

```python
# Тренування моделі
history = model.fit(
    datagen.flow(x_train, y_train, batch_size=32),
    validation_data=(x_test, y_test),
    epochs=10,
    steps_per_epoch=len(x_train) // 32
)
```

####Крок 3: Оцінка моделі

Після навчання ми оцінюємо модель на тестових даних:

```python
# Оцінка на тестовому наборі
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Точність на тестовому наборі: {accuracy * 100:.2f}%")
```

####Крок 4: Візуалізація результатів

Відобразимо графіки точності та втрат під час навчання:

```python
import matplotlib.pyplot as plt

# Графік точності
plt.plot(history.history['accuracy'], label='Точність на тренуванні')
plt.plot(history.history['val_accuracy'], label='Точність на валідації')
plt.title('Точність моделі')
plt.xlabel('Епоха')
plt.ylabel('Точність')
plt.legend()
plt.show()

# Графік втрат
plt.plot(history.history['loss'], label='Втрати на тренуванні')
plt.plot(history.history['val_loss'], label='Втрати на валідації')
plt.title('Втрати моделі')
plt.xlabel('Епоха')
plt.ylabel('Втрати')
plt.legend()
plt.show()
```

**Оцінка результатів та графіки**

Після навчання моделі ResNet50 на датасеті CIFAR-100 були отримані наступні результати:

*Точність на тестовому наборі:*

- Тестова точність склала 0.99%, що свідчить про те, що модель ще не адаптована до датасету через низьку узгодженість між тренувальними та тестовими результатами.

*Втрати та точність під час навчання:*

- Графіки показують стабільно низьку точність як на тренуванні, так і на валідації (~0.01).

- Втрати залишаються практично незмінними після першої епохи, що вказує на необхідність тоншої настройки моделі.

Графік точності моделі:

### Етап 4: Адаптація моделі до задачі класифікації CIFAR-100

```python
# Розморожуємо верхні шари
for layer in base_model.layers[-30:]:  # Розморожуємо останні 30 шарів
    layer.trainable = True
```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.0001),  # Зменшили на порядок
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_accuracy',
                               patience=5,  # Зупиняємо, якщо 5 епох немає прогресу
                               restore_best_weights=True)

history = model.fit(
    datagen.flow(x_train, y_train, batch_size=32),
    validation_data=(x_test, y_test),
    epochs=5,
    callbacks=[early_stopping],
    steps_per_epoch=len(x_train) // 32
)
```
